```python
import nltk
nltk.download('punkt_tab')
nltk.download('wordnet')
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('words')
nltk.download('maxent_ne_chunker_tab')
```

```python
text = "Hello I am Aditya. I am a final year Undergraduate student at Alliance University. I am taking this NLP course."
print(text)
statement = sent_tokenize(text)
print(statement)
```

```python
print(statement)
stemmer = PorterStemmer()
words = []
for sentence in statement:
  words = words + word_tokenize(sentence)
  words = [stemmer.stem(word) for word in words]
  sentence = ' '.join(words)
print(sentence)
```

```python
lemmatizer = WordNetLemmatizer()
for sentence in statement:
  newwords =  [lemmatizer.lemmatize(word) for word in words]
  sentence = ' '.join(newwords)
print(sentence)
```

```python
tag_word = nltk.pos_tag(words)
print(tag_word)
```

```python
result = nltk.ne_chunk(tag_word)
print(result)
```
