```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
text = "Hello I am Aditya. I am taking this NLP course."
print(text)
statement = sent_tokenize(text)
print(statement)
```

```python
print(statement)
words_sen1 = word_tokenize(statement[0])
words_sen2 = word_tokenize(statement[1])
words = words_sen1 + words_sen2
print(words)
```
