```python
!pip install spacy
!pip install pyspellchecker
import spacy
import en_core_web_sm
import nltk
import re
from spellchecker import SpellChecker
from nltk.stem import PorterStemmer
spacy.load("en_core_web_sm")
nlp = en_core_web_sm.load()
stemmer = PorterStemmer()
```

```python
doc = nlp("Hello, I am Aditya. I am taking this NLP Course.")
print(doc)
sentences = [sentence.text.lower() for sentence in doc.sents]
print(sentences)
```

```python
words = [word.text.lower() for word in doc]
print(words)
```

```python
stemmed_words = [stemmer.stem(word.text).lower() for word in doc]
print(stemmed_words)
```

```python
processed_words = [word.lemma_.lower() for word in doc]
for i in range(len(words)):
  print(words[i] + ",", processed_words[i])
```

```python
tagged_words = [word.pos_ for word in doc]
for i in range(len(words)):
  print(words[i] + ", ", tagged_words[i] )
```

```python
test = nlp("The MacBook Pro is the best Laptop in the world. It is 12% more efficient than any other laptop. Apple outdid itself this year")
low = [word.label_ for word in test.ents]
print(low)
```

```python
filtered_words = [ word.text for word in doc if not word.is_stop]
print(filtered_words)
```

```python
quote = """Ohhhhh sorry. I am runnnnnning late for classsssssss!!!!!!!! I didn't bump into on purposeeee! """
sample_text = nlp(quote)
sentences = [sentence.text.lower() for sentence in sample_text.sents]
change = ""
for sentence in sentences:
  change += " " + re.sub(r'(.)\1+', r'\1', sentence)
change = change.strip()
sample_text = nlp(change)
spell_checker = SpellChecker()
corrected_words = []
for word in sample_text:
  if word.is_alpha:
    correct = spell_checker.correction(word.text)
    corrected_words.append(correct)
  else:
    corrected_words.append(word.text)
corrected_sentences = " "
for word in corrected_words:
  if word in r"\b[.!?,] | n't | 'll\b":
    corrected_sentences += word
  else:
    corrected_sentences += " " + word
corrected_sentences = corrected_sentences.strip()
sample_text = nlp(corrected_sentences)
print(sample_text)
```
