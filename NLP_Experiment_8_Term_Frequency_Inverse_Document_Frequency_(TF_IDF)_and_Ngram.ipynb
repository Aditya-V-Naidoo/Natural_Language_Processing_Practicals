```python
speech = sent_tokenize(paragraph.lower())
print(speech)
```

```python
pattern_punctuation = r'\W'
speech = [re.sub(pattern_punctuation, ' ', x) for x in speech]
print(speech)
```

```python
word2count = {}
stop_words = set(stopwords.words('english'))
for line in speech:
  words  = word_tokenize(line)
  filtered_words = [word for word in words if word.lower() not in stop_words]
  for word in filtered_words:
    if word not in word2count.keys():
      word2count[word] = 1
    else:
      word2count[word] += 1
print(word2count)
```

```python
most_freq = heapq.nlargest(50,word2count,key = word2count.get)
print(most_freq)
```

```python
TF_IDF = {}
total_word_count = 0
total_sentences  = len(speech)
for line in speech:
  words  = word_tokenize(line)
  total_word_count += len(words)
for word in most_freq:
    if word not in TF_IDF.keys():
      TF_IDF[word] = 0
    else:
      continue
```

```python
sent_count = {}
for word in TF_IDF.keys():
  sent_count[word] = 0
  for data in speech:
    if word in word_tokenize(data):
      sent_count[word] += 1
    else:
      continue
```

```python
for key in TF_IDF.keys():
  TF_IDF[key] = (word2count[key]/total_word_count) * math.log(total_sentences/sent_count[key])

print(TF_IDF)
```

```python
uni_grams = list(ngrams(word2count.keys(), 1))
print(uni_grams)
bi_grams = list(ngrams(word2count.keys(), 2))
print(bi_grams)
tri_grams = list(ngrams(word2count.keys(), 3))
print(tri_grams)
```
