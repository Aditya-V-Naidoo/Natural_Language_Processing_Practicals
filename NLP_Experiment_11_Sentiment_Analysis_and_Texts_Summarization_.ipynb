```python
import bs4 as bs
import urllib.request
import re
import requests
import lxml
import nltk
import heapq
from nltk.corpus import stopwords
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
url = "https://en.wikipedia.org/wiki/Global_warming"

headers = {"User-Agent": "Safari/605.1.15"}

response = requests.get(url, headers= headers)

source = response.text
```

```python
soup = bs.BeautifulSoup(source, 'lxml')

text = ""
for paragraph in soup.find_all('p'):
  text += paragraph.text

text = re.sub(r'\b[0-9]\b', ' ', text)
text = re.sub(r'\W', ' ', text)
text = text.lower()
text = re.sub(r'\d', ' ', text)
text = re.sub(r'\s+', ' ', text)
```

```python
sentences = nltk.sent_tokenize(text)
sentences = [nltk.word_tokenize(sentence) for sentence in sentences]

stop_words = set(stopwords.words('english'))
sentences = [[word for word in sentence if word not in stop_words] for sentence in sentences]
```

```python
word2count = {}
for sentence in sentences:
  for word in sentence:
    if word not in word2count.keys():
      word2count[word] = 1
    else:
      word2count[word] += 1
```

```python
for key in word2count.keys():
  word2count[key] = word2count[key]/max(word2count.values())

print(word2count)
```

```python
sent2score = {}
sentences = nltk.sent_tokenize(text)

for sentence in sentences:
  for word in nltk.word_tokenize(sentence):
    if word in word2count.keys():
      if len(sentence.split(" ")) < 25:
        if sentence not in sent2score:
          sent2score[sentence] = word2count[word]
        else:
          sent2score[sentence] += word2count[word]
```

```python
best_sentences = heapq.nlargest(5, sent2score, key=sent2score.get)


for sentence in best_sentences:
  print(sentence)
```
